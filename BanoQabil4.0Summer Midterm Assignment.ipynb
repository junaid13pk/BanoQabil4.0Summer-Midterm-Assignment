```python
import tkinter as tk
from tkinter import simpledialog, messagebox
import random

# Step 1: List of random welcome messages (quotes/jokes/motivational lines)
welcome_messages = [
    "🌟 'The only limit to our realization of tomorrow is our doubts of today.' – Franklin D. Roosevelt",
    "😂 Why don’t programmers like nature? It has too many bugs!",
    "💪 'Success is not final, failure is not fatal: It is the courage to continue that counts.' – Winston Churchill",
    "🎯 'Your time is limited, so don’t waste it living someone else’s life.' – Steve Jobs",
    "😎 'In the middle of every difficulty lies opportunity.' – Albert Einstein",
    "🤣 Why don’t skeletons fight each other? They don’t have the guts.",
    "🙌 'Happiness depends upon ourselves.' – Aristotle",
    "😂 Why did the computer go to the doctor? Because it caught a virus!",
    "🌱 'Do what you can, with what you have, where you are.' – Theodore Roosevelt",
    "😄 Why did the math book look sad? Because it had too many problems.",
    "🚀 'The best way to predict the future is to invent it.' – Alan Kay",
    "🤣 Why don’t scientists trust atoms? Because they make up everything!",
    "🌸 'Happiness is not something ready-made. It comes from your own actions.' – Dalai Lama",
    "🎉 Why don’t oysters donate to charity? Because they are shellfish.",
    "🔥 'Don’t watch the clock; do what it does. Keep going.' – Sam Levenson",
    "😂 Parallel lines have so much in common. It’s a shame they’ll never meet.",
    "☀️ 'Keep your face always toward the sunshine—and shadows will fall behind you.' – Walt Whitman",
    "🤣 Why did the scarecrow win an award? Because he was outstanding in his field.",
    "🏆 'Hardships often prepare ordinary people for an extraordinary destiny.' – C.S. Lewis",
    "😂 Why did the bicycle fall over? Because it was two-tired!",
    "🌈 'Believe you can and you’re halfway there.' – Theodore Roosevelt",
    "🤣 Why can’t your nose be 12 inches long? Because then it would be a foot.",
    "💡 'It always seems impossible until it’s done.' – Nelson Mandela",
    "😂 Why did the golfer bring two pairs of pants? In case he got a hole in one.",
    "🎯 'Opportunities don’t happen. You create them.' – Chris Grosser",
    "😂 Why don’t eggs tell jokes? They’d crack each other up.",
    "🤣 Why did the tomato blush? Because it saw the salad dressing!",
    "😆 Why did the computer go to art school? Because it wanted to learn how to draw a better screen.",
    "😂 Why don’t fish play basketball? Because they’re afraid of the net.",
    "🤣 Why did the coffee file a police report? It got mugged!",
    "😄 Why can’t your hand be 12 inches long? Because then it would be a foot!",
    "😂 Why don’t cows ever have any money? Because farmers milk them dry!",
    "🤣 Why was the broom late? It swept in!",
    "😂 Why don’t seagulls fly over the bay? Because then they’d be bagels!",
    "🤣 Why was the computer cold? Because it forgot to close its Windows!",
    "😂 Why did the man put his money in the blender? Because he wanted to make liquid assets!",
    "🤣 Why don’t some couples go to the gym? Because some relationships don’t work out!",
    "😂 Why did the chicken join a band? Because it had the drumsticks!",
]

# Step 2: Pre-saved messages for specific names
friendship_messages = {
    "junaid": "Keep shining like the star you are",
    "ali": "Your friendship means the world to me",
    "ahsan": "Stay strong and keep smiling",
    "fatima": "You are an inspiration to many",
    "aquib": "Your kindness brings light into every room",
    "shaukat": "Your wisdom and positivity inspire everyone around you",
}

# Step 3: Emoji dictionary for translation
emoji_dict = {
    "keep": "🤝", "shining": "✨", "star": "🌟", "friendship": "💖", "world": "🌍",
    "smiling": "😊", "strong": "💪", "inspiration": "🌈", "many": "👥", "you": "👉",
    "me": "🙋", "light": "💡", "positivity": "☀️", "wisdom": "🧠", "kindness": "🤲",
    "happy": "😄", "love": "❤️", "laugh": "😂", "life": "🌱", "dream": "💭",
    "success": "🏆", "goal": "🎯", "hope": "🕊️", "smile": "😁", "peace": "☮️",
    "friend": "👫", "team": "🤝", "family": "👨‍👩‍👧‍👦", "heart": "💓", "strength": "🦾",
    "shine": "🔆", "focus": "🎯", "power": "⚡", "future": "🚀", "journey": "🛤️",
    "challenge": "⚔️", "victory": "🥇", "brave": "🦁", "smart": "🧠", "beautiful": "🌸",
    "cool": "😎", "energy": "🔋", "leader": "👑", "win": "🏅",
}

def translate_to_emojis(message):
    """Convert words in the message into emojis using emoji_dict"""
    words = message.lower().split()
    result = []
    for word in words:
        if word in emoji_dict:
            result.append(emoji_dict[word])
        else:
            result.append(word)
    return " ".join(result)

# Step 4: Create main Tkinter window (hidden for now)
root = tk.Tk()
root.withdraw()

# Show random welcome message popup
random_message = random.choice(welcome_messages)
messagebox.showinfo("Welcome", random_message)

# Step 5: Ask user for their name
user_name = simpledialog.askstring("Your Name", "Please enter your name:")

if user_name:
    formatted_name = user_name.strip().capitalize()

    # Personalized greeting
    messagebox.showinfo("Greeting", f"Welcome to the Fun Friendship Capsule, Mr./Ms. {formatted_name}!")

    lower_name = formatted_name.lower()
    if lower_name in friendship_messages:
        personal_msg = friendship_messages[lower_name]
        emoji_msg = translate_to_emojis(personal_msg)
        messagebox.showinfo(
            "Special Message",
            f"Mr./Ms. {formatted_name} has a message for you:\n\n{personal_msg}\n\nEmoji Version:\n{emoji_msg}"
        )
    else:
        no_msg = f"Sorry, there is no saved message for you, Mr./Ms. {formatted_name}."
        emoji_msg = translate_to_emojis("Stay happy and keep smiling always")
        messagebox.showwarning("No Message", f"{no_msg}\n\nBut here’s something for you:\n{emoji_msg}")

    # Step 6: Option to hear jokes or quotes
    while True:
        ask_quote = messagebox.askyesno(
            "Fun Capsule", f"{formatted_name}, would you like to hear some jokes or quotes?"
        )
        if ask_quote:
            another_message = random.choice(welcome_messages)
            messagebox.showinfo("For You", another_message)
        else:
            break

    # Step 7: Thank you and personalized goodbye
    messagebox.showinfo("Goodbye", f"🙏 Thank you for playing the Fun Friendship Capsule, {formatted_name}!\nGoodbye 👋")

root.destroy()
```
